```python
# you can use various libraries to view images
img = mpimg.imread('cameraman.tif')
plt.figure()
imgplot = plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Lets make some operations and save the results

# change the data type of the image
img2 = img.astype('float64')

# Normalize the image
out = (img2 - np.min(img2))/(np.max(img2) - np.min(img2))
img_org = out.copy();

path = 'cameraman2.png'
plt.imsave(path, img, cmap='gray')

# increase the pixel intensity values in the image
out = out + 0.3
out[out > 1.0] = 1.0
out[out < 0.0] = 0.0

out_img = np.uint8(out*255)

plt.figure()
imgplot = plt.imshow(out_img, cmap='gray', vmin=0, vmax=255)
plt.show()

minVal = np.min(out_img)
maxVal = np.max(out_img)

# write the image to memory
path = 'cameraman_bright.png'
plt.imsave(path, out_img, cmap='gray', vmin=0,vmax=255)
#plt.savefig('cameraman_bright.png')
```
